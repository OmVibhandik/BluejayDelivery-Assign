```python
import pandas as pd   # Import pandas for data manipulation
from datetime import timedelta   # Import timedelta for time calculations
```

```python
data =  pd.read_excel('Assignment_Timecard.xlsx') # Load timecard data from Excel file
```

```python
data.columns  # View column names
data = data.drop(['Unnamed: 9','Unnamed: 10'],axis=1) # Drop unnecessary columns
```

```python
data.head(20) # Preview the first 20 rows
```

```python
# Define a function to check for seven consecutive days of work
dic = {}

for index, row in data.iterrows():
    key = (row['Position ID'], row['Employee Name'])
    if key not in dic:
        dic[key] = []
    dic[key].append(row)
    # print(row['Position ID'])
```

```python

# Define a function to check for seven consecutive days of work
def check_seven_consecutive_days(date_list):
    if pd.isnull(date_list[0]):
        return False
    date_ints = [date.toordinal() for date in date_list]
    date_ints.sort()  # Sort chronologically
    counter = 1 # Initialize a counter
    prev_date = date_ints[0]
    
    for date in date_ints[1:]:
        if date - prev_date == 1:  # Check for consecutive days
            counter += 1
        else:
            counter = 1
            
        if counter >= 7:  # Found seven consecutive days
            return True
        
        prev_date = date
    return False
```

```python
# Identify employees who worked seven consecutive days
list_of_people_working_seven_consecutive_days = []

for i in dic:
    date_list = [row['Time'] for row in dic[i]]
    if check_seven_consecutive_days(date_list):
        list_of_people_working_seven_consecutive_days.append(i)
```

```python
# Print results for consecutive days
print("List of people who has worked seven consecutive days")
print(len(list_of_people_working_seven_consecutive_days))
list_of_people_working_seven_consecutive_days
```

```python
# Identify employees with short breaks between shifts
list_of_people_from_one_to_ten = []
for i in dic:
    date_list = [row['Time'] for row in dic[i]]
    for j in range(len(date_list)-1):
        diff = date_list[j+1] - date_list[j]
        diff = (diff.seconds / 60)/60 # Convert timedelta to hours
        if 1 < diff < 10:
            list_of_people_from_one_to_ten.append(i)
            break
```

```python
# Print results for short breaks
print("List of people who has taken less than 10 hours but more than 1 hours between shifts")
print(len(list_of_people_from_one_to_ten))
list_of_people_from_one_to_ten
```

```python
# Identify employees who worked long shifts
list_of_people_working_14_hrs = []
for i in dic:
    in_date = [row['Time'] for row in dic[i]]
    out_date = [row['Time Out'] for row in dic[i]]
    if pd.isnull(in_date[0]):
        continue    # Skip if missing data
    for j in range(len(in_date)):
        diff = out_date[j] - in_date[j]
        diff = (diff.seconds / 60)/60  
        if diff > 14:
            list_of_people_working_14_hrs.append(i)
            break
```

```python
# Print results for long shifts
print("List of people who has worked for more than 14 hours in a single shift")
print(len(list_of_people_working_14_hrs))
list_of_people_working_14_hrs
```
